<h1> Chapter 2: Fourier Analysis of Signals</h1> 
</div>

<br/>

<p>
The Fourier transform is perhaps the most fundamental tool in signal processing. Chapter 2 of <a href="http://www.music-processing.de">[Müller, FMP, Springer 2015]</a> approaches Fourier analysis from various perspectives and discusses their most important properties in the context of audio signal processing. In particular, the notion of a spectrogram, which yields a time&ndash;frequency representation of an audio signal, is introduced.    
</p>
