<h1> Basics</h1> 
</div>

<br/>

<p>
The notebooks of <strong>Part B</strong> serve different purposes. First, these notebooks describe the main tools used for developing the FMP notebooks. Second, they give short introductions of the relevant technical concepts while providing links to more detailed tutorials. Third, the notebooks give examples for best practices for generating and using code, figures, and sound elements.         
</p>

<div class="alert alert-block alert-warning">
<strong>Note:</strong> The notebooks of the Part B do not have to be read in the order they are specified. The B-part notebooks serve as a documentation and reference of overarching aspects that concern the FMP notebooks of the subsequent parts.
<br>
</div>
