<h1> Chapter 6: Tempo and Beat Tracking</h1> 
</div>

<br/>

<p>
Tempo and beat are further fundamental properties of music. In Chapter 6 of <a href="http://www.music-processing.de">[Müller, FMP, Springer 2015]</a>, we introduce the basic ideas on how to extract tempo-related information from audio recordings. In this scenario, a first challenge is to locate note onset information&mdash;a task that requires methods for detecting changes in energy and spectral content. To derive tempo and beat information, note onset candidates are then analyzed with regard to quasiperiodic patterns. This leads us to the study of general methods for local periodicity analysis of time series.     
</p>
